```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = np.arange(-1.0,1.0,0.2)
x2 = np.arange(-1.0,1.0,0.2)
# 출력값을 저장할 10x10 그리드
Z = np.zeros((10,10))

#각각에 대한 가중치 값 고정
w_x1 = 2.5
w_x2 = 3.0
# 편향
bias = 0.1 #다음 뉴런으로 도움을 보내기 위한 영향을 끼침
# 그리드맵의 각 그리드별 뉴런의 연산
for i in range(10):
    for j in range(10):
        # 입력과 가중치 곱의 합 + 편향
        u = x1[i] * w_x1 + x2[j] * w_x2 + bias
        y = 1/ (1+ np.exp(-u))# 시그모이드 함수
        Z[j][i] = y
        
plt.imshow(Z,"gray", vmin = 0.0, vmax = 1.0)
plt.colorbar()
plt.xlabel("X1")
plt.ylabel("X2")
plt.show()
```

## 4.다중신경망

```python
#하나이상의 뉴럴네트워크를 사용하여 만듯것이 딥러닝
```

```python
#X,Y는 입력피쳐
X = np.arange(-1.0, 1.0, 0.2)
Y = np.arange(-1.0, 1.0, 0.2)

Z = np.zeros((10,10))

#히든레이어로 넘어가는 가중치
w_im = np.array([[4.0,4.0],[4.0,4.0]])#인풋에서 미들로
w_mo = np.array([[1.0],[-1.0]])#미들에서 아웃풋으로갈때
#레이어와레이어 넘어갈때 쉐입신경쓰기
b_im = np.array([3.0, -3.0])#인풋에서 미들로
b_mo = np.array([0.1])

def middle_layer(x, w, b):
    u = np.dot(x, w) + b
    return 1 / (1+ np.exp(-u))# 시그모이드 함수에 넣어 리턴

def output_layer(x, w, b):
    u = np.dot(x, w) + b
    return u

for i in range(10):
    for j in range(10):
        inp = np.array([X[i], Y[j]])
        mid = middle_layer(inp, w_im, b_im)
        out = output_layer(mid, w_mo, b_mo)#출력 
        
        Z[j][i] = out[0]#넘파이 배열로 나오므 0번으로 설정해야한다
        
plt.imshow(Z, 'gray', vmin=0.0, vmax=1.0)
plt.colorbar()
plt.xlabel('Fearture X')
plt.ylabel('Fearture Y')
plt.show()
```

## 분류

```python
X = np.arange(-1.0, 1.0, 0.1)
Y = np.arange(-1.0, 1.0, 0.1)

w_im = np.array([[1.0, 2.0],[2.0,3.0]])
w_mo = np.array([[-1.0, 1.0], [1.0, -1.0]])

b_im = np.array([0.3, -0.3])
b_mo = np.array([0.4, 0.1])

def middle_layer(x,w,b):
    u = np.dot(x, w) + b
    return 1 / (1 + np.exp(-u))

def output_layer(x,w,b):
    u = np.dot(x,w) + b
    return np.exp(u) / np.sum(np.exp(u))

x_1 = []
y_1 = []
x_2 = []
y_2 = []

for i in range(20):
    for j in range(20):
        inp = np.array([X[i], Y[j]])
        mid = middle_layer(inp, w_im, b_im)
        out = output_layer(mid, w_mo, b_mo)
        #마약야ㅔ 출력 0번째가 1번째 보다 크다면 x_1에 어펜드 하기             
        
        if out[0] > out[1]:
            x_1.append(X[i])
            y_1.append(Y[j])
        else:
            x_2.append(X[i])
            y_2.append(Y[j])
                
                
plt.scatter(x_1, y_1, marker='+')
plt.scatter(x_2, y_2, marker='o')
plt.show()# 바이너리 분류
```
